# Importing Libraries

```python
import numpy as Num
import pandas as Pand
import matplotlib.pyplot as plt
import seaborn as SeaBo
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

# Importing Datset

```python
data_set=Pand.read_csv("wine_set.csv")
data_set.head()
```

# Finding whether missing values exist or not

```python
data_set.isnull().sum()
```

# statistical measures of dataset

```python
data_set.describe()
```

# Number of values for each quality

```python
SeaBo.catplot(x="quality",data=data_set,kind="count")
```

# Histogram for features

```python
data_set.hist(bins=25,figsize=[10,10])
plt.show()
```

# Constructing the heatmap to determine the correlation between the features

```python
correlation=data_set.corr()
plt.figure(figsize=(10,10))
SeaBo.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

# Dependent and Independent variables

```python
x=data_set.iloc[:,:-1].values
y=data_set.iloc[:,-1].values
print(x)
print(y)
```

# Train_Test_Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Using Linear Regression Model

```python
Line_Reg=LinearRegression()
Line_Reg.fit(x_train,y_train)
```

# Providing test data

```python
test_data=Line_Reg.predict(x_test)
print(test_data)
```

# Calculation Evaluation Metrics

```python
mse=mean_squared_error(y_test,test_data)
rmse=Num.sqrt(mse)
r2=r2_score(y_test,test_data)
mae=mean_absolute_error(y_test,test_data)
print("The R_squared score:",r2)
print("The Root mean squared error:",rmse)
print("The Mean absolute error:",mae)
```

# Visualization the distribution of differences between actual and predicted values

```python
SeaBo.distplot([y_test-test_data],bins=50)
```
